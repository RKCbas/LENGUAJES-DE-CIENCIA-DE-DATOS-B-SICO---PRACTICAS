16.  Utiliza un bucle while para encontrar el menor número divisible por todos los números del 1 al 10.

```r
numero <- 1
encontrado <- FALSE
while (!encontrado) {
  divisible <- TRUE
  for (i in 1:10) {
    if (numero %% i != 0) {
      divisible <- FALSE
      break
    }
  }
  if (divisible) {
    encontrado <- TRUE
  } else {
    numero <- numero + 1
  }
}
cat("El menor número divisible por todos del 1 al 10 es:", numero)
```

20. Escribe un bucle while que lea números del usuario hasta que ingresen un 0, luego imprime la suma de esos números.

```r
nums <- c()
numero <- -1

while (numero != 0) {
  numero <- as.numeric(readline("Ingresa un número (0 para terminar): "))

  if (numero != 0) {
    nums <- c(nums, numero)
  }
}

cat("Números ingresados:\n")
cat(nums)

cat("\nLa suma es:", sum(nums), "\n")
```

22.  Escribe un bucle repeat que encuentre el factorial de un número. Utiliza break para terminar el bucle cuando sea apropiado.

```r
numero <- sample(1:6, 1)
factorial <- 1
i <- 1
repeat {
  factorial <- factorial * i
  i <- i + 1
  if (i > numero) {
    break
  }
}
cat("Factorial de", numero, "es", factorial)
```

23.  Crea un bucle repeat que solicite al usuario números y los sume hasta que el usuario ingrese un número negativo.

```r
nums_entrada <- c()
suma <- 0

repeat {
  numero <- as.numeric(readline("Ingresa un número (negativo para terminar): "))

  if (numero < 0) {
    break
  }

  nums_entrada <- c(nums_entrada, numero)
  suma <- suma + numero
}

cat("Los números ingresados fueron:\n")
cat(nums_entrada)

cat("\nLa suma es:", suma, "\n")
```

24.  Utiliza un bucle repeat para generar números aleatorios hasta obtener un número mayor que 0.95, luego imprime el número de intentos realizados.

```r
intentos <- 0
repeat {
  numero <- runif(1)
  intentos <- intentos + 1
  if (numero > 0.95) {
    break
  }
}
cat("Número de intentos:", intentos)
cat("\nNúmero obtenido:", numero)
```

25.  Implementa un bucle repeat que itere hasta que la diferencia entre dos estimaciones consecutivas de pi (usando la serie de Leibniz) sea menor a un umbral específico. La serie de Leibniz para el cálculo de se define como:

![25_image](image.png)

```r
pi_estimado <- 0
pi_anterior <- 0
umbral <- 0.0001
n <- 0
repeat {
  pi_anterior <- pi_estimado
  termino <- ((-1)^n) / (2 * n + 1)
  pi_estimado <- pi_estimado + termino
  n <- n + 1
  if (abs(4 * pi_estimado - 4 * pi_anterior) < umbral && n > 1) {
    break
  }
}
cat("Estimación de pi:", 4 * pi_estimado)
cat("\nIteraciones:", n)
```

30.  Implementa un bucle anidado para calcular el producto cartesiano de dos conjuntos almacenados en vectores.

```r
conjunto1 <- c(1, 2, 3)
conjunto2 <- c("X", "Y")
producto_cartesiano <- list()
k <- 1
for (i in conjunto1) {
  for (j in conjunto2) {
    producto_cartesiano[[k]] <- c(i, j)
    k <- k + 1
  }
}
print(producto_cartesiano)
```

## Manejo de datos

1. Utiliza lapply para convertir todos los caracteres en una lista de cadenas a mayúsculas.

```r
lista_cadenas <- list("hola", "mundo", "r", "programación")
uppercase <- lapply(lista_cadenas, toupper)
print(uppercase)
```

2. Escribe un script que use sapply para calcular la media de cada vector en una lista.

```r
lista_vectores <- list(c(1, 2, 3, 4, 5), c(10, 20, 30), c(5, 15, 25, 35))
medias <- sapply(lista_vectores, mean)
print(medias)
```

3. Utiliza vapply para verificar si los elementos en una lista de vectores son todos positivos, devolviendo un vector lógico.

```r
lista_nums <- list(c(1, 2, 3), c(-1, 2, 3), c(5, 10, 15))
todos_positivos <- vapply(lista_nums, function(x) all(x > 0), logical(1))
print(todos_positivos)
```

4. Implementa lapply para dividir cada número en una lista de vectores por su máximo valor.

```r
lista_datos <- list(c(10, 20, 30, 40), c(5, 15, 25), c(100, 200, 300))
nums <- lapply(lista_datos, function(x) x / max(x))
print(nums)
```

5. Usa sapply para contar cuántos elementos únicos hay en cada vector de una lista.

```r
lista_con_duplicados <- list(c(1, 1, 2, 3, 3), c(5, 5, 5, 10), c(1, 2, 3, 4, 5))
unique <- sapply(lista_con_duplicados, function(x) length(unique(x)))
print(unique)
```

6. Utiliza apply para sumar los valores en cada fila de una matriz.

```r
matriz <- matrix(1:12, nrow = 3, ncol = 4)
print("Matriz:")
print(matriz)
suma_filas <- apply(matriz, 1, sum)
print("Suma por fila:")
print(suma_filas)
```

7. Escribe un script que use apply para normalizar (restar la media y dividir por la desviación estándar) las columnas de una matriz.

```r
matriz <- matrix(rnorm(20), nrow = 5, ncol = 4)
print("Matriz original:")
print(matriz)
matriz_normalizada <- apply(matriz, 2, function(x) (x - mean(x)) / sd(x))
print("Matriz normalizada:")
print(matriz_normalizada)
```

10. Crea una función personalizada y úsala con apply para calcular el rango (máximo - mínimo) de los valores en las filas de una matriz

```r
calcular_rango <- function(x) {
  max(x) - min(x)
}
matriz <- matrix(c(1, 5, 10, 2, 8, 15, 3, 12, 20), nrow = 3, ncol = 3)
print("Matriz:")
print(matriz)
rangos <- apply(matriz, 1, calcular_rango)
print("Rango por fila:")
print(rangos)
```
