## 1. Piedra Papel, Tijera, Lagarto y Spock

Crea un programa que calcule quien gana mÃ¡s partidas al piedra,
papel, tijera, lagarto, spock.
- El resultado puede ser: "Player 1", "Player 2", "Tie" (empate)
- La funciÃ³n recibe un listado que contiene pares, representando cada jugada.
- El par puede contener combinaciones de "ğŸ—¿" (piedra), "ğŸ“„" (papel),
  "âœ‚ï¸" (tijera), "ğŸ¦" (lagarto) o "ğŸ––" (spock).
- Ejemplo. Entrada: [("ğŸ—¿","âœ‚ï¸"), ("âœ‚ï¸","ğŸ—¿"), ("ğŸ“„","âœ‚ï¸")]. Resultado: "Player 2".
- Debes buscar informaciÃ³n sobre cÃ³mo se juega con estas 5 posibilidades.

```python
import os
import getpass

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

def show_rules():
    print("""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘        PIEDRA, PAPEL, TIJERA, LAGARTO, SPOCK              â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘  âœ‚ï¸  Tijera   â†’ corta Papel, decapita Lagarto             â•‘
â•‘  ğŸ“„ Papel    â†’ cubre Piedra, refuta Spock                 â•‘
â•‘  ğŸ—¿ Piedra   â†’ aplasta Tijera, aplasta Lagarto            â•‘
â•‘  ğŸ¦ Lagarto  â†’ envenena Spock, come Papel                 â•‘
â•‘  ğŸ–– Spock    â†’ rompe Tijera, vaporiza Piedra              â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """)

def get_choice(player_name):
    options = {
        '1': ('ğŸ—¿', 'Piedra'),
        '2': ('ğŸ“„', 'Papel'),
        '3': ('âœ‚ï¸', 'Tijera'),
        '4': ('ğŸ¦', 'Lagarto'),
        '5': ('ğŸ––', 'Spock')
    }
    
    print(f"\nğŸ® Turno de {player_name}")
    print("-" * 30)
    for k, v in options.items():
        print(f"  {k}. {v[0]} {v[1]}")
    print("-" * 30)
    
    while True:
        # getpass oculta la entrada para que el otro jugador no vea
        choice = getpass.getpass(f"Elige (1-5) [oculto]: ")
        if choice in options:
            return options[choice][0]
        print("âŒ OpciÃ³n invÃ¡lida. Intenta de nuevo.")

def get_winner(p1, p2):
    wins = {
        "ğŸ—¿": ["âœ‚ï¸", "ğŸ¦"],
        "ğŸ“„": ["ğŸ—¿", "ğŸ––"],
        "âœ‚ï¸": ["ğŸ“„", "ğŸ¦"],
        "ğŸ¦": ["ğŸ“„", "ğŸ––"],
        "ğŸ––": ["âœ‚ï¸", "ğŸ—¿"]
    }
    
    if p1 == p2:
        return 0  # Empate
    elif p2 in wins[p1]:
        return 1  # Gana P1
    return 2  # Gana P2

def play_game():
    clear_screen()
    show_rules()
    
    # ConfiguraciÃ³n
    while True:
        try:
            rounds = int(input("Â¿CuÃ¡ntas rondas quieren jugar? "))
            if rounds > 0:
                break
            print("Debe ser un nÃºmero positivo.")
        except ValueError:
            print("Ingresa un nÃºmero vÃ¡lido.")
    
    p1_name = input("Nombre del Jugador 1: ") or "Jugador 1"
    p2_name = input("Nombre del Jugador 2: ") or "Jugador 2"
    
    p1_score, p2_score = 0, 0
    history = []
    
    for r in range(1, rounds + 1):
        clear_screen()
        print(f"\n{'='*40}")
        print(f"         RONDA {r} de {rounds}")
        print(f"  {p1_name}: {p1_score} | {p2_name}: {p2_score}")
        print(f"{'='*40}")
        
        p1_choice = get_choice(p1_name)
        clear_screen()
        p2_choice = get_choice(p2_name)
        
        result = get_winner(p1_choice, p2_choice)
        history.append((p1_choice, p2_choice))
        
        clear_screen()
        print(f"\nğŸ¯ RESULTADO RONDA {r}")
        print(f"  {p1_name}: {p1_choice}  vs  {p2_choice} :{p2_name}")
        
        if result == 0:
            print("  ğŸ¤ Â¡EMPATE!")
        elif result == 1:
            print(f"  ğŸ† Â¡Gana {p1_name}!")
            p1_score += 1
        else:
            print(f"  ğŸ† Â¡Gana {p2_name}!")
            p2_score += 1
        
        if r < rounds:
            input("\nPresiona Enter para continuar...")
    
    # Resultado final
    print(f"\n{'='*40}")
    print("         ğŸ RESULTADO FINAL ğŸ")
    print(f"{'='*40}")
    print(f"  {p1_name}: {p1_score} victorias")
    print(f"  {p2_name}: {p2_score} victorias")
    print(f"{'='*40}")
    
    if p1_score > p2_score:
        print(f"\nğŸ‰ Â¡{p1_name} GANA EL JUEGO! ğŸ‰")
    elif p2_score > p1_score:
        print(f"\nğŸ‰ Â¡{p2_name} GANA EL JUEGO! ğŸ‰")
    else:
        print("\nğŸ¤ Â¡EL JUEGO TERMINA EN EMPATE!")

if __name__ == "__main__":
    while True:
        play_game()
        again = input("\nÂ¿Jugar de nuevo? (s/n): ").lower()
        if again != 's':
            print("\nğŸ‘‹ Â¡Hasta pronto!")
            break
```

## 2. Generador de contraseÃ±as

Escribe un programa que sea capaz de generar contraseÃ±as de forma aleatoria.
PodrÃ¡s configurar generar contraseÃ±as con los siguientes parÃ¡metros:
- Longitud: Entre 8 y 16.
- Con o sin letras mayÃºsculas.
- Con o sin nÃºmeros.
- Con o sin sÃ­mbolos.
(Pudiendo combinar todos estos parÃ¡metros entre ellos)

```python
import random
import string

def generate_password(
    length: int = 12,
    uppercase: bool = True,
    numbers: bool = True,
    symbols: bool = True
) -> str:
    """
    Genera una contraseÃ±a aleatoria con los parÃ¡metros especificados.
    
    Args:
        length: Longitud de la contraseÃ±a (8-16)
        uppercase: Incluir letras mayÃºsculas
        numbers: Incluir nÃºmeros
        symbols: Incluir sÃ­mbolos especiales
    
    Returns:
        ContraseÃ±a generada
    """
    # Validar longitud
    if not 8 <= length <= 16:
        raise ValueError("La longitud debe estar entre 8 y 16")
    
    # Construir conjunto de caracteres
    chars = string.ascii_lowercase  # Siempre incluir minÃºsculas
    required = [random.choice(string.ascii_lowercase)]
    
    if uppercase:
        chars += string.ascii_uppercase
        required.append(random.choice(string.ascii_uppercase))
    
    if numbers:
        chars += string.digits
        required.append(random.choice(string.digits))
    
    if symbols:
        syms = "!@#$%^&*()_+-=[]{}|;:,.<>?"
        chars += syms
        required.append(random.choice(syms))
    
    # Generar resto de caracteres
    remaining = length - len(required)
    password = required + [random.choice(chars) for _ in range(remaining)]
    
    # Mezclar para que los requeridos no estÃ©n siempre al inicio
    random.shuffle(password)
    
    return ''.join(password)


def interactive_generator():
    """Interfaz interactiva para generar contraseÃ±as."""
    print("\n" + "="*50)
    print("    ğŸ” GENERADOR DE CONTRASEÃ‘AS SEGURAS ğŸ”")
    print("="*50)
    
    # Longitud
    while True:
        try:
            length = int(input("\nğŸ“ Longitud (8-16): "))
            if 8 <= length <= 16:
                break
            print("âŒ Debe estar entre 8 y 16")
        except ValueError:
            print("âŒ Ingresa un nÃºmero vÃ¡lido")
    
    # Opciones
    def ask_yes_no(prompt):
        while True:
            resp = input(prompt).strip().lower()
            if resp in ['s', 'si', 'sÃ­', 'y', 'yes', '']:
                return True
            if resp in ['n', 'no']:
                return False
            print("âŒ Responde 's' o 'n'")
    
    uppercase = ask_yes_no("ğŸ”  Â¿Incluir mayÃºsculas? (S/n): ")
    numbers = ask_yes_no("ğŸ”¢ Â¿Incluir nÃºmeros? (S/n): ")
    symbols = ask_yes_no("ğŸ”£ Â¿Incluir sÃ­mbolos? (S/n): ")
    
    # Cantidad
    while True:
        try:
            qty = int(input("\nğŸ“ Â¿CuÃ¡ntas contraseÃ±as generar? (1-10): "))
            if 1 <= qty <= 10:
                break
            print("âŒ Debe estar entre 1 y 10")
        except ValueError:
            print("âŒ Ingresa un nÃºmero vÃ¡lido")
    
    # Generar y mostrar
    print("\n" + "-"*50)
    print("ğŸ² CONTRASEÃ‘AS GENERADAS:")
    print("-"*50)
    
    for i in range(qty):
        pwd = generate_password(length, uppercase, numbers, symbols)
        print(f"  {i+1}. {pwd}")
    
    print("-"*50)
    print(f"âœ… Longitud: {length} | MayÃºsculas: {'SÃ­' if uppercase else 'No'} | "
          f"NÃºmeros: {'SÃ­' if numbers else 'No'} | SÃ­mbolos: {'SÃ­' if symbols else 'No'}")


if __name__ == "__main__":
    while True:
        interactive_generator()
        again = input("\nÂ¿Generar mÃ¡s contraseÃ±as? (s/n): ").lower()
        if again not in ['s', 'si', 'sÃ­', 'y', 'yes']:
            print("\nğŸ‘‹ Â¡Hasta pronto!")
            break
```

## 3. Adivina la palabra

Crea un pequeÃ±o juego que consista en adivinar palabras en un nÃºmero mÃ¡ximo de intentos:
- El juego comienza proponiendo una palabra aleatoria incompleta
  - Por ejemplo "m_ur_d_v", y el nÃºmero de intentos que le quedan
- El usuario puede introducir Ãºnicamente una letra o una palabra (de la misma longitud que
  la palabra a adivinar)
  - Si escribe una letra y acierta, se muestra esa letra en la palabra. Si falla, se resta
    uno al nÃºmero de intentos
  - Si escribe una resoluciÃ³n y acierta, finaliza el juego, en caso contrario, se resta uno
    al nÃºmero de intentos
  - Si el contador de intentos llega a 0, el jugador pierde
- La palabra debe ocultar de forma aleatoria letras, y nunca puede comenzar ocultando mÃ¡s del 60%
- Puedes utilizar las palabras que quieras y el nÃºmero de intentos que consideres

```python
import random

# Lista de palabras para el juego
WORDS = [
    "python", "programacion", "algoritmo", "variable", "funcion",
    "biblioteca", "desarrollo", "computadora", "teclado", "pantalla",
    "internet", "servidor", "cliente", "archivo", "carpeta",
    "sistema", "memoria", "procesador", "software", "hardware",
    "elefante", "mariposa", "dinosaurio", "cocodrilo", "murcielago",
    "estrella", "universo", "galaxia", "planeta", "aventura"
]


def hide_letters(word: str, max_hidden_percent: float = 0.6) -> tuple[str, set]:
    """
    Oculta letras aleatorias de una palabra.
    
    Args:
        word: Palabra a ocultar
        max_hidden_percent: Porcentaje mÃ¡ximo de letras a ocultar (default 60%)
    
    Returns:
        Tupla con (palabra oculta, conjunto de Ã­ndices ocultos)
    """
    length = len(word)
    max_hidden = int(length * max_hidden_percent)
    num_hidden = random.randint(1, max_hidden)
    
    # Seleccionar Ã­ndices aleatorios para ocultar
    hidden_indices = set(random.sample(range(length), num_hidden))
    
    # Crear palabra oculta
    hidden_word = ""
    for i, char in enumerate(word):
        hidden_word += "_" if i in hidden_indices else char
    
    return hidden_word, hidden_indices


def display_status(hidden_word: str, attempts: int, guessed_letters: set):
    """Muestra el estado actual del juego."""
    print("\n" + "="*50)
    print(f"  Palabra: {' '.join(hidden_word)}")
    print(f"  Intentos restantes: {'â¤ï¸ ' * attempts}({attempts})")
    if guessed_letters:
        print(f"  Letras usadas: {', '.join(sorted(guessed_letters))}")
    print("="*50)


def play_game():
    """Ejecuta una partida del juego."""
    # Seleccionar palabra aleatoria
    word = random.choice(WORDS).lower()
    hidden_word, hidden_indices = hide_letters(word)
    
    # ConfiguraciÃ³n
    attempts = 6
    guessed_letters = set()
    current_display = list(hidden_word)
    
    print("\n" + "*"*50)
    print("    ğŸ¯ JUEGO DE ADIVINAR PALABRAS ğŸ¯")
    print("*"*50)
    print("\nğŸ“‹ REGLAS:")
    print("  â€¢ Escribe una LETRA para revelar posiciones")
    print("  â€¢ Escribe la PALABRA COMPLETA para resolver")
    print("  â€¢ Si fallas, pierdes un intento")
    print(f"\nğŸ’¡ La palabra tiene {len(word)} letras")
    
    while attempts > 0:
        display_status(''.join(current_display), attempts, guessed_letters)
        
        # Verificar si ya se adivinÃ³
        if '_' not in current_display:
            print("\nğŸ‰ Â¡FELICIDADES! Â¡Has adivinado la palabra!")
            print(f"   La palabra era: {word.upper()}")
            return True
        
        # Obtener entrada del usuario
        guess = input("\nğŸ‘‰ Tu letra o palabra: ").lower().strip()
        
        if not guess:
            print("âŒ Debes escribir algo.")
            continue
        
        # Caso: Una letra
        if len(guess) == 1:
            if not guess.isalpha():
                print("âŒ Solo se permiten letras.")
                continue
            
            if guess in guessed_letters:
                print(f"âš ï¸  Ya usaste la letra '{guess}'. Prueba otra.")
                continue
            
            guessed_letters.add(guess)
            
            # Verificar si la letra estÃ¡ en la palabra
            if guess in word:
                # Revelar todas las ocurrencias de la letra
                found = False
                for i, char in enumerate(word):
                    if char == guess and current_display[i] == '_':
                        current_display[i] = guess
                        found = True
                
                if found:
                    print(f"âœ… Â¡Bien! La letra '{guess}' estÃ¡ en la palabra.")
                else:
                    print(f"â„¹ï¸  La letra '{guess}' ya estaba visible.")
            else:
                attempts -= 1
                print(f"âŒ La letra '{guess}' no estÃ¡ en la palabra.")
        
        # Caso: Palabra completa
        elif len(guess) == len(word):
            if guess == word:
                print("\nğŸ‰ Â¡EXCELENTE! Â¡Adivinaste la palabra!")
                print(f"   La palabra era: {word.upper()}")
                return True
            else:
                attempts -= 1
                print(f"âŒ '{guess}' no es la palabra correcta.")
        
        else:
            print(f"âŒ La palabra debe tener {len(word)} letras, no {len(guess)}.")
    
    # Sin intentos
    print("\n" + "="*50)
    print("ğŸ’€ Â¡GAME OVER! Te quedaste sin intentos.")
    print(f"   La palabra era: {word.upper()}")
    print("="*50)
    return False


def main():
    """FunciÃ³n principal del juego."""
    wins = 0
    losses = 0
    
    while True:
        result = play_game()
        if result:
            wins += 1
        else:
            losses += 1
        
        print(f"\nğŸ“Š EstadÃ­sticas: {wins} victorias | {losses} derrotas")
        
        again = input("\nÂ¿Jugar otra vez? (s/n): ").lower().strip()
        if again not in ['s', 'si', 'sÃ­', 'y', 'yes']:
            print("\nğŸ‘‹ Â¡Gracias por jugar! Â¡Hasta pronto!")
            break


if __name__ == "__main__":
    main()
```

Crea un programa que detecte cuando el famoso "CÃ³digo Konami" se ha introducido correctamente desde el teclado. Si sucede esto, debe notificarse mostrando un mensaje en la terminal.

```python
"""
Detector del CÃ³digo Konami
Secuencia: â†‘ â†‘ â†“ â†“ â† â†’ â† â†’ B A

Teclas en el teclado:
- Flechas: â†‘ â†“ â† â†’
- Letras: B, A
"""

try:
    from pynput import keyboard
except ImportError:
    print("Instalando dependencia necesaria...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'pynput'])
    from pynput import keyboard

# Secuencia del CÃ³digo Konami
KONAMI_CODE = [
    'up', 'up', 'down', 'down', 
    'left', 'right', 'left', 'right', 
    'b', 'a'
]

# Estado actual de la secuencia
current_sequence = []


def show_welcome():
    """Muestra mensaje de bienvenida."""
    print("\n" + "="*60)
    print("    ğŸ® DETECTOR DEL CÃ“DIGO KONAMI ğŸ®")
    print("="*60)
    print("\nğŸ“‹ Secuencia a introducir:")
    print("   â†‘ â†‘ â†“ â†“ â† â†’ â† â†’ B A")
    print("\nğŸ¹ Usa las flechas del teclado y las teclas B y A")
    print("   Presiona ESC para salir")
    print("\n" + "-"*60)
    print("ğŸ” Esperando entrada...\n")


def show_easter_egg():
    """Muestra el mensaje secreto al completar el cÃ³digo."""
    print("\n" + "*"*60)
    print("""
    â–ˆâ–ˆâ•—  â–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ•—   â–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ•—   â–ˆâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•—
    â–ˆâ–ˆâ•‘ â–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•”â•â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ•—  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•—â–ˆâ–ˆâ–ˆâ–ˆâ•— â–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘
    â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â• â–ˆâ–ˆâ•‘   â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â–ˆâ–ˆâ•— â–ˆâ–ˆâ•‘â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â–ˆâ–ˆâ–ˆâ–ˆâ•”â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘
    â–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•— â–ˆâ–ˆâ•‘   â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ•—â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•”â•â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘â•šâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘
    â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•—â•šâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ•”â•â–ˆâ–ˆâ•‘ â•šâ–ˆâ–ˆâ–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘  â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘ â•šâ•â• â–ˆâ–ˆâ•‘â–ˆâ–ˆâ•‘
    â•šâ•â•  â•šâ•â• â•šâ•â•â•â•â•â• â•šâ•â•  â•šâ•â•â•â•â•šâ•â•  â•šâ•â•â•šâ•â•     â•šâ•â•â•šâ•â•
    """)
    print("    ğŸ‰ Â¡CÃ“DIGO KONAMI ACTIVADO! ğŸ‰")
    print("    +30 VIDAS DESBLOQUEADAS")
    print("*"*60 + "\n")


def get_key_name(key):
    """Convierte una tecla a su nombre normalizado."""
    try:
        # Teclas especiales (flechas)
        if key == keyboard.Key.up:
            return 'up'
        elif key == keyboard.Key.down:
            return 'down'
        elif key == keyboard.Key.left:
            return 'left'
        elif key == keyboard.Key.right:
            return 'right'
        elif key == keyboard.Key.esc:
            return 'esc'
        # Teclas de caracteres
        elif hasattr(key, 'char') and key.char:
            return key.char.lower()
    except AttributeError:
        pass
    return None


def display_progress():
    """Muestra el progreso visual de la secuencia."""
    symbols = {
        'up': 'â†‘', 'down': 'â†“', 'left': 'â†', 'right': 'â†’',
        'b': 'B', 'a': 'A'
    }
    
    progress = [symbols.get(k, k) for k in current_sequence]
    remaining = len(KONAMI_CODE) - len(current_sequence)
    
    print(f"   Secuencia: {' '.join(progress)}", end="")
    if remaining > 0:
        print(f" ... ({remaining} restantes)")
    else:
        print()


def on_press(key):
    """Callback para cada tecla presionada."""
    global current_sequence
    
    key_name = get_key_name(key)
    
    # Salir con ESC
    if key_name == 'esc':
        print("\nğŸ‘‹ Â¡Hasta pronto!")
        return False
    
    # Ignorar teclas no vÃ¡lidas
    if key_name not in ['up', 'down', 'left', 'right', 'b', 'a']:
        return True
    
    # Verificar si la tecla es la siguiente esperada
    expected_index = len(current_sequence)
    
    if expected_index < len(KONAMI_CODE) and key_name == KONAMI_CODE[expected_index]:
        current_sequence.append(key_name)
        display_progress()
        
        # Verificar si se completÃ³ el cÃ³digo
        if len(current_sequence) == len(KONAMI_CODE):
            show_easter_egg()
            current_sequence = []  # Reiniciar para poder hacerlo de nuevo
            print("ğŸ”„ Puedes intentarlo de nuevo o presionar ESC para salir\n")
    else:
        # Secuencia incorrecta, reiniciar
        if current_sequence:
            print("   âŒ Secuencia incorrecta. Reiniciando...")
        current_sequence = []
        
        # Verificar si la tecla presionada es el inicio de una nueva secuencia
        if key_name == KONAMI_CODE[0]:
            current_sequence.append(key_name)
            display_progress()
    
    return True


def main():
    """FunciÃ³n principal."""
    show_welcome()
    
    # Escuchar eventos de teclado
    with keyboard.Listener(on_press=on_press) as listener:
        listener.join()


if __name__ == "__main__":
    main()
```

 Crea una funciÃ³n que transforme las pulsaciones del T9 a su

- Debes buscar cuÃ¡l era su correspondencia original.
- Cada bloque de pulsaciones va separado por un guiÃ³n.

    Entrada: 6-666-88-777-33-3-33-888

```python
"""
Decodificador T9 Multi-tap
Convierte pulsaciones de teclado numÃ©rico a letras.

Correspondencia T9:
    1: .    2: ABC   3: DEF
    4: GHI  5: JKL   6: MNO
    7: PQRS 8: TUV   9: WXYZ
    0: (espacio)
"""

# Mapeo T9: cada nÃºmero corresponde a un grupo de letras
T9_MAP = {
    '1': '.',
    '2': 'ABC',
    '3': 'DEF',
    '4': 'GHI',
    '5': 'JKL',
    '6': 'MNO',
    '7': 'PQRS',
    '8': 'TUV',
    '9': 'WXYZ',
    '0': ' '
}


def decode_t9(input_code: str) -> str:
    """
    Decodifica una secuencia T9 a texto.
    
    Args:
        input_code: Secuencia de pulsaciones separadas por guiones
                   Ejemplo: "6-666-88-777-33-3-33-888"
    
    Returns:
        Texto decodificado en mayÃºsculas
    
    Raises:
        ValueError: Si el formato es invÃ¡lido
    """
    if not input_code or not input_code.strip():
        return ""
    
    # Separar bloques por guiones
    blocks = input_code.strip().split('-')
    result = []
    
    for block in blocks:
        block = block.strip()
        
        if not block:
            continue
        
        # Validar que todos los caracteres sean el mismo nÃºmero
        if not block.isdigit():
            raise ValueError(f"Bloque invÃ¡lido: '{block}'. Solo se permiten nÃºmeros.")
        
        if len(set(block)) > 1:
            raise ValueError(f"Bloque invÃ¡lido: '{block}'. Todos los dÃ­gitos deben ser iguales.")
        
        digit = block[0]
        count = len(block)
        
        if digit not in T9_MAP:
            raise ValueError(f"DÃ­gito no vÃ¡lido: '{digit}'")
        
        letters = T9_MAP[digit]
        
        # Calcular Ã­ndice (con ciclo si se excede)
        index = (count - 1) % len(letters)
        result.append(letters[index])
    
    return ''.join(result)


def encode_t9(text: str) -> str:
    """
    Codifica texto a secuencia T9.
    
    Args:
        text: Texto a codificar
    
    Returns:
        Secuencia T9 separada por guiones
    """
    # Crear mapeo inverso
    reverse_map = {}
    for digit, letters in T9_MAP.items():
        for i, letter in enumerate(letters):
            reverse_map[letter] = digit * (i + 1)
    
    result = []
    for char in text.upper():
        if char in reverse_map:
            result.append(reverse_map[char])
        elif char == ' ':
            result.append('0')
        # Ignorar caracteres no mapeables
    
    return '-'.join(result)


def show_keyboard():
    """Muestra el teclado T9."""
    print("""
    â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
    â•‘        TECLADO T9 MULTI-TAP       â•‘
    â• â•â•â•â•â•â•â•â•â•â•â•â•¦â•â•â•â•â•â•â•â•â•â•â•â•¦â•â•â•â•â•â•â•â•â•â•â•â•£
    â•‘  1  .     â•‘  2  ABC   â•‘  3  DEF   â•‘
    â• â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•£
    â•‘  4  GHI   â•‘  5  JKL   â•‘  6  MNO   â•‘
    â• â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•£
    â•‘  7  PQRS  â•‘  8  TUV   â•‘  9  WXYZ  â•‘
    â• â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•¬â•â•â•â•â•â•â•â•â•â•â•â•£
    â•‘     *     â•‘  0  â£     â•‘     #     â•‘
    â•šâ•â•â•â•â•â•â•â•â•â•â•â•©â•â•â•â•â•â•â•â•â•â•â•â•©â•â•â•â•â•â•â•â•â•â•â•â•
    """)


def interactive_mode():
    """Modo interactivo para decodificar/codificar."""
    print("\n" + "="*50)
    print("    ğŸ“± DECODIFICADOR T9 MULTI-TAP ğŸ“±")
    print("="*50)
    
    show_keyboard()
    
    print("ğŸ“‹ INSTRUCCIONES:")
    print("  â€¢ Cada bloque de pulsaciones se separa con '-'")
    print("  â€¢ Un bloque debe contener el mismo nÃºmero repetido")
    print("  â€¢ Ejemplo: 6-666-88-777-33-3-33-888 = MOUREDEV")
    print("\n" + "-"*50)
    
    while True:
        print("\nÂ¿QuÃ© deseas hacer?")
        print("  1. Decodificar T9 â†’ Texto")
        print("  2. Codificar Texto â†’ T9")
        print("  3. Salir")
        
        option = input("\nOpciÃ³n: ").strip()
        
        if option == '1':
            code = input("\nğŸ”¢ Introduce cÃ³digo T9: ").strip()
            try:
                result = decode_t9(code)
                print(f"\nCÃ³digo: {code}")
                print(f"\nâœ… Resultado: {result}")
            except ValueError as e:
                print(f"\nâŒ Error: {e}")
        
        elif option == '2':
            text = input("\nğŸ“ Introduce texto: ").strip()
            result = encode_t9(text)
            print(f"\nTexto: {text}")
            print(f"\nâœ… CÃ³digo T9: {result}")
        
        elif option == '3':
            print("\nğŸ‘‹ Â¡Hasta pronto!")
            break
        
        else:
            print("âŒ OpciÃ³n no vÃ¡lida")


# Ejemplos y pruebas
if __name__ == "__main__":
    # Pruebas
    print("="*50)
    print("         PRUEBAS DE DECODIFICACIÃ“N")
    print("="*50)
    
    tests = [
        ("6-666-88-777-33-3-33-888", "MOUREDEV"),
        ("4-666-555-2", "HOLA"),
        ("7-999-8-44-666-66", "PYTHON"),
        ("2-3-33-222", "ADEC"),
    ]
    
    for code, expected in tests:
        result = decode_t9(code)
        status = "âœ…" if result == expected else "âŒ"
        print(f"{status} {code}")
        print(f"   Esperado: {expected} | Obtenido: {result}\n")
    
    # Modo interactivo
    interactive_mode()
```

# 6. La carrera de coches

Crea un programa que simule la competiciÃ³n de dos coches en una pista.
- Los dos coches estarÃ¡n representados por ğŸš™ y ğŸš—. Y la meta por ğŸ.
- Cada pista tendrÃ¡ entre 1 y 3 Ã¡rboles ğŸŒ² colocados de forma aleatoria.
- Las dos pistas tendrÃ¡n una longitud configurable de guiones bajos "_".
- Los coches comenzarÃ¡n en la parte derecha de las pistas. Ejemplo:
  ğŸ____ğŸŒ²_____ğŸš™
  ğŸ_ğŸŒ²____ğŸŒ²___ğŸš—

El juego se desarrolla por turnos de forma automÃ¡tica, y cada segundo
se realiza una acciÃ³n sobre los coches (moviÃ©ndose a la vez), hasta que
uno de ellos (o los dos a la vez) llega a la meta.
- Acciones:
  - Avanzar entre 1 a 3 posiciones hacia la meta.
  - Si al avanzar, el coche finaliza en la posiciÃ³n de un Ã¡rbol,
    se muestra ğŸ’¥ y no avanza durante un turno.
  - Cada turno se imprimen las pistas y sus elementos.
  - Cuando la carrera finalice, se muestra el coche ganador o el empate.

```python
import random
import time
import os


class Car:
    """Representa un coche en la carrera."""
    
    def __init__(self, name: str, emoji: str, track_length: int):
        self.name = name
        self.emoji = emoji
        self.position = track_length  # Empieza al final
        self.track_length = track_length
        self.stunned = False  # Si chocÃ³ con un Ã¡rbol
        self.finished = False
    
    def move(self, trees: set) -> tuple[int, bool]:
        """
        Mueve el coche.
        
        Returns:
            tuple: (posiciones avanzadas, chocÃ³ con Ã¡rbol)
        """
        if self.stunned:
            self.stunned = False
            return 0, False
        
        if self.finished:
            return 0, False
        
        # Avanzar entre 1 y 3 posiciones
        advance = random.randint(1, 3)
        new_position = max(0, self.position - advance)
        
        # Verificar si llegÃ³ a la meta
        if new_position == 0:
            self.position = 0
            self.finished = True
            return advance, False
        
        # Verificar si chocÃ³ con un Ã¡rbol
        if new_position in trees:
            self.position = new_position
            self.stunned = True
            return advance, True
        
        self.position = new_position
        return advance, False


class RaceTrack:
    """Representa una pista de carreras."""
    
    def __init__(self, length: int = 20):
        self.length = length
        self.trees = self._place_trees()
    
    def _place_trees(self) -> set:
        """Coloca entre 1 y 3 Ã¡rboles aleatoriamente."""
        num_trees = random.randint(1, 3)
        # Los Ã¡rboles no pueden estar en la posiciÃ³n 0 (meta) ni en la Ãºltima (inicio)
        available_positions = list(range(1, self.length))
        return set(random.sample(available_positions, min(num_trees, len(available_positions))))
    
    def display(self, car: Car) -> str:
        """Genera la visualizaciÃ³n de la pista."""
        track = ['_'] * (self.length + 1)
        track[0] = 'ğŸ'
        
        # Colocar Ã¡rboles
        for tree_pos in self.trees:
            track[tree_pos] = 'ğŸŒ²'
        
        # Colocar coche
        if car.stunned and car.position in self.trees:
            track[car.position] = 'ğŸ’¥'
        elif car.position == 0:
            track[car.position] = car.emoji + 'ğŸ'
        else:
            track[car.position] = car.emoji
        
        return ''.join(track)


def clear_screen():
    """Limpia la pantalla."""
    os.system('cls' if os.name == 'nt' else 'clear')


def display_race(car1: Car, track1: RaceTrack, car2: Car, track2: RaceTrack, turn: int):
    """Muestra el estado actual de la carrera."""
    clear_screen()
    print("\n" + "="*60)
    print(f"{'ğŸ CARRERA DE COCHES ğŸ':^60}")
    print("="*60)
    print(f"\nğŸ”„ Turno: {turn}\n")
    print(f"{car1.name} {car1.emoji}")
    print(track1.display(car1))
    print()
    print(f"{car2.name} {car2.emoji}")
    print(track2.display(car2))
    print("\n" + "-"*60)


def show_action(car: Car, advance: int, crashed: bool):
    """Muestra la acciÃ³n realizada por un coche."""
    if crashed:
        print(f"ğŸ’¥ {car.name} chocÃ³ con un Ã¡rbol y queda aturdido!")
    elif advance > 0:
        print(f"ğŸƒ {car.name} avanza {advance} posiciÃ³n(es)")
    elif car.stunned:
        print(f"ğŸ˜µ {car.name} estÃ¡ recuperÃ¡ndose del choque")
    elif car.finished:
        print(f"ğŸ‰ {car.name} mantiene su posiciÃ³n en la meta")


def race_game(track_length: int = 20):
    """Ejecuta la carrera de coches."""
    # Crear pistas
    track1 = RaceTrack(track_length)
    track2 = RaceTrack(track_length)
    
    # Crear coches
    car1 = Car("Coche 1", "ğŸš™", track_length)
    car2 = Car("Coche 2", "ğŸš—", track_length)
    
    turn = 0
    
    # Mostrar estado inicial
    display_race(car1, track1, car2, track2, turn)
    print("\nğŸ Â¡La carrera estÃ¡ por comenzar!")
    input("\nPresiona Enter para iniciar...")
    
    # Loop principal
    while not (car1.finished or car2.finished):
        turn += 1
        time.sleep(1)  # Esperar 1 segundo
        
        # Mover coches
        advance1, crashed1 = car1.move(track1.trees)
        advance2, crashed2 = car2.move(track2.trees)
        
        # Mostrar estado
        display_race(car1, track1, car2, track2, turn)
        
        # Mostrar acciones
        show_action(car1, advance1, crashed1)
        show_action(car2, advance2, crashed2)
        
        # Verificar fin de carrera
        if car1.finished or car2.finished:
            break
    
    # Mostrar resultado
    time.sleep(1)
    print("\n" + "="*60)
    print(f"{'ğŸ† RESULTADO FINAL ğŸ†':^60}")
    print("="*60)
    
    if car1.finished and car2.finished:
        print("\nğŸ¤ Â¡EMPATE! Ambos coches llegaron a la meta")
    elif car1.finished:
        print(f"\nğŸ† Â¡{car1.name} {car1.emoji} GANA LA CARRERA!")
    else:
        print(f"\nğŸ† Â¡{car2.name} {car2.emoji} GANA LA CARRERA!")
    
    print(f"\nğŸ“Š Turnos totales: {turn}")
    print("="*60)


def main():
    """FunciÃ³n principal."""
    print("\n" + "="*60)
    print(f"{'ğŸ CARRERA DE COCHES ğŸ':^60}")
    print("="*60)
    print("\nğŸï¸  Dos coches compiten en pistas con obstÃ¡culos")
    print("ğŸŒ² Los Ã¡rboles pueden hacer que los coches choquen")
    print("ğŸ¯ El primero en llegar a la meta gana\n")
    
    while True:
        try:
            length = int(input("ğŸ“ Longitud de la pista (10-50, recomendado 20): "))
            if 10 <= length <= 50:
                break
            print("âŒ La longitud debe estar entre 10 y 50")
        except ValueError:
            print("âŒ Ingresa un nÃºmero vÃ¡lido")
    
    while True:
        race_game(length)
        
        again = input("\nÂ¿Otra carrera? (s/n): ").lower().strip()
        if again not in ['s', 'si', 'sÃ­', 'y', 'yes']:
            print("\nğŸ‘‹ Â¡Gracias por jugar!")
            break


if __name__ == "__main__":
    main()
```
