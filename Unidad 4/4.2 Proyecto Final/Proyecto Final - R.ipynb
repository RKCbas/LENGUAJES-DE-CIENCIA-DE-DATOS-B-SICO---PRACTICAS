## 1. Piedra Papel, Tijera, Lagarto y Spock

Crea un programa que calcule quien gana m√°s partidas al piedra,
papel, tijera, lagarto, spock.
- El resultado puede ser: "Player 1", "Player 2", "Tie" (empate)
- La funci√≥n recibe un listado que contiene pares, representando cada jugada.
- El par puede contener combinaciones de "üóø" (piedra), "üìÑ" (papel),
  "‚úÇÔ∏è" (tijera), "ü¶é" (lagarto) o "üññ" (spock).
- Ejemplo. Entrada: [("üóø","‚úÇÔ∏è"), ("‚úÇÔ∏è","üóø"), ("üìÑ","‚úÇÔ∏è")]. Resultado: "Player 2".
- Debes buscar informaci√≥n sobre c√≥mo se juega con estas 5 posibilidades.

```r
clear_screen <- function() {
  if (.Platform$OS.type == "windows") shell("cls") else system("clear")
}

show_rules <- function() {
  cat("
+===========================================================+
|        PIEDRA, PAPEL, TIJERA, LAGARTO, SPOCK              |
+===========================================================+
|  1. Piedra   -> aplasta Tijera, aplasta Lagarto           |
|  2. Papel    -> cubre Piedra, refuta Spock                |
|  3. Tijera   -> corta Papel, decapita Lagarto             |
|  4. Lagarto  -> envenena Spock, come Papel                |
|  5. Spock    -> rompe Tijera, vaporiza Piedra             |
+===========================================================+
\n")
}

get_choice <- function(player_name) {
  options <- list(
    "1" = c("Piedra", "1"),
    "2" = c("Papel", "2"),
    "3" = c("Tijera", "3"),
    "4" = c("Lagarto", "4"),
    "5" = c("Spock", "5")
  )

  cat(paste0("\n>>> Turno de ", player_name, " <<<\n"))
  cat("------------------------------\n")
  cat("  1. Piedra\n")
  cat("  2. Papel\n")
  cat("  3. Tijera\n")
  cat("  4. Lagarto\n")
  cat("  5. Spock\n")
  cat("------------------------------\n")
  cat("(El otro jugador debe mirar hacia otro lado)\n")

  repeat {
    choice <- readline(prompt = paste0(player_name, ", elige (1-5): "))
    if (choice %in% names(options)) {
      # Limpiar pantalla para ocultar la elecci√≥n
      invisible(readline(prompt = "Presiona Enter y pasa el turno..."))
      clear_screen()
      return(choice)
    }
    cat("Opci√≥n invalida. Intenta de nuevo.\n")
  }
}

get_winner <- function(p1, p2) {
  # Retorna: 0 = empate, 1 = gana P1, 2 = gana P2
  wins <- list(
    "1" = c("3", "4"),  # Piedra vence a Tijera y Lagarto
    "2" = c("1", "5"),  # Papel vence a Piedra y Spock
    "3" = c("2", "4"),  # Tijera vence a Papel y Lagarto
    "4" = c("2", "5"),  # Lagarto vence a Papel y Spock
    "5" = c("3", "1")   # Spock vence a Tijera y Piedra
  )

  if (p1 == p2) return(0)
  if (p2 %in% wins[[p1]]) return(1)
  2
}

get_name <- function(choice) {
  names <- c("1" = "Piedra", "2" = "Papel", "3" = "Tijera",
             "4" = "Lagarto", "5" = "Spock")
  names[choice]
}

get_emoji <- function(choice) {
  emojis <- c("1" = "[PIEDRA]", "2" = "[PAPEL]", "3" = "[TIJERA]",
              "4" = "[LAGARTO]", "5" = "[SPOCK]")
  emojis[choice]
}

play_game <- function() {
  clear_screen()
  show_rules()

  # Configuration
  rounds <- 0
  while (rounds <= 0) {
    rounds_input <- readline(prompt = "Cuantas rondas quieren jugar? ")
    rounds <- suppressWarnings(as.integer(rounds_input))
    if (is.na(rounds) || rounds <= 0) {
      cat("Ingresa un numero positivo.\n")
      rounds <- 0
    }
  }

  p1_name <- readline(prompt = "Nombre del Jugador 1: ")
  if (p1_name == "") p1_name <- "Jugador 1"

  p2_name <- readline(prompt = "Nombre del Jugador 2: ")
  if (p2_name == "") p2_name <- "Jugador 2"

  p1_score <- 0
  p2_score <- 0

  for (r in 1:rounds) {
    clear_screen()
    cat(paste0("\n", strrep("=", 40), "\n"))
    cat(paste0("         RONDA ", r, " de ", rounds, "\n"))
    cat(paste0(
      "  ", p1_name, ": ", p1_score, " | ", p2_name, ": ", p2_score, "\n"
    ))
    cat(paste0(strrep("=", 40), "\n"))

    p1_choice <- get_choice(p1_name)
    p2_choice <- get_choice(p2_name)

    result <- get_winner(p1_choice, p2_choice)

    clear_screen()
    cat(paste0("\n>>> RESULTADO RONDA ", r, " <<<\n"))
    cat(paste0("  ", p1_name, ": ", get_emoji(p1_choice),
               "  vs  ", get_emoji(p2_choice), " :", p2_name, "\n"))

    if (result == 0) {
      cat("  *** EMPATE! ***\n")
    } else if (result == 1) {
      cat(paste0("  *** Gana ", p1_name, "! ***\n"))
      p1_score <- p1_score + 1
    } else {
      cat(paste0("  *** Gana ", p2_name, "! ***\n"))
      p2_score <- p2_score + 1
    }

    if (r < rounds) {
      invisible(readline(prompt = "\nPresiona Enter para continuar..."))
    }
  }

  # Resultado final
  cat(paste0("\n", strrep("=", 40), "\n"))
  cat("         RESULTADO FINAL\n")
  cat(paste0(strrep("=", 40), "\n"))
  cat(paste0("  ", p1_name, ": ", p1_score, " victorias\n"))
  cat(paste0("  ", p2_name, ": ", p2_score, " victorias\n"))
  cat(paste0(strrep("=", 40), "\n"))

  if (p1_score > p2_score) {
    cat(paste0("\n*** ", p1_name, " GANA EL JUEGO! ***\n"))
  } else if (p2_score > p1_score) {
    cat(paste0("\n*** ", p2_name, " GANA EL JUEGO! ***\n"))
  } else {
    cat("\n*** EL JUEGO TERMINA EN EMPATE! ***\n")
  }
}

# Ejecutar el juego
main <- function() {
  repeat {
    play_game()
    again <- readline(prompt = "\nJugar de nuevo? (s/n): ")
    if (tolower(again) != "s") {
      cat("\nHasta pronto!\n")
      break
    }
  }
}

# Iniciar
main()
```

## 2. Generador de contrase√±as

Escribe un programa que sea capaz de generar contrase√±as de forma aleatoria.
Podr√°s configurar generar contrase√±as con los siguientes par√°metros:
- Longitud: Entre 8 y 16.
- Con o sin letras may√∫sculas.
- Con o sin n√∫meros.
- Con o sin s√≠mbolos.
(Pudiendo combinar todos estos par√°metros entre ellos)

```r
generate_password <- function(length = 12,
                              uppercase = TRUE,
                              numbers = TRUE,
                              symbols = TRUE) {
  #' Genera una contrase√±a aleatoria con los par√°metros especificados.
  #' @param length Longitud de la contrase√±a (8-16)
  #' @param uppercase Incluir letras may√∫sculas
  #' @param numbers Incluir n√∫meros
  #' @param symbols Incluir s√≠mbolos especiales
  #' @return Contrase√±a generada

  # Validar longitud
  if (length < 8 || length > 16) {
    stop("La longitud debe estar entre 8 y 16")
  }

  # Definir conjuntos de caracteres
  lowercase_chars <- letters
  uppercase_chars <- LETTERS
  number_chars <- 0:9
  symbol_chars <- unlist(strsplit("!@#$%^&*()_+-=[]{}|;:,.<>?", ""))

  # Construir conjunto de caracteres y requeridos
  chars <- lowercase_chars
  required <- sample(lowercase_chars, 1)

  if (uppercase) {
    chars <- c(chars, uppercase_chars)
    required <- c(required, sample(uppercase_chars, 1))
  }

  if (numbers) {
    chars <- c(chars, number_chars)
    required <- c(required, sample(number_chars, 1))
  }

  if (symbols) {
    chars <- c(chars, symbol_chars)
    required <- c(required, sample(symbol_chars, 1))
  }

  # Generar resto de caracteres
  remaining <- length - length(required)
  password <- c(required, sample(chars, remaining, replace = TRUE))

  # Mezclar para que los requeridos no est√©n siempre al inicio
  password <- sample(password)

  return(paste(password, collapse = ""))
}


interactive_generator <- function() {
  #' Interfaz interactiva para generar contrase√±as.

  cat("\n")
  cat(strrep("=", 50), "\n")
  cat("    GENERADOR DE CONTRASE√ëAS SEGURAS\n")
  cat(strrep("=", 50), "\n")

  # Longitud
  length <- 0
  while (length < 8 || length > 16) {
    length_input <- readline(prompt = "\nLongitud (8-16): ")
    length <- suppressWarnings(as.integer(length_input))
    if (is.na(length) || length < 8 || length > 16) {
      cat("Error: Debe estar entre 8 y 16\n")
      length <- 0
    }
  }

  # Funci√≥n auxiliar para preguntas s√≠/no
  ask_yes_no <- function(prompt) {
    while (TRUE) {
      resp <- tolower(trimws(readline(prompt = prompt)))
      if (resp %in% c("s", "si", "s√≠", "y", "yes", "")) {
        return(TRUE)
      }
      if (resp %in% c("n", "no")) {
        return(FALSE)
      }
      cat("Error: Responde 's' o 'n'\n")
    }
  }

  uppercase <- ask_yes_no("Incluir may√∫sculas? (S/n): ")
  numbers <- ask_yes_no("Incluir n√∫meros? (S/n): ")
  symbols <- ask_yes_no("Incluir s√≠mbolos? (S/n): ")

  # Cantidad
  qty <- 0
  while (qty < 1 || qty > 10) {
    qty_input <- readline(prompt = "\nCuantas contrase√±as generar? (1-10): ")
    qty <- suppressWarnings(as.integer(qty_input))
    if (is.na(qty) || qty < 1 || qty > 10) {
      cat("Error: Debe estar entre 1 y 10\n")
      qty <- 0
    }
  }

  # Generar y mostrar
  cat("\n")
  cat(strrep("-", 50), "\n")
  cat("CONTRASE√ëAS GENERADAS:\n")
  cat(strrep("-", 50), "\n")

  for (i in 1:qty) {
    pwd <- generate_password(length, uppercase, numbers, symbols)
    cat(sprintf("  %d. %s\n", i, pwd))
  }

  cat(strrep("-", 50), "\n")
  cat(sprintf("Longitud: %d | May√∫sculas: %s | N√∫meros: %s | S√≠mbolos: %s\n",
              length,
              ifelse(uppercase, "Si", "No"),
              ifelse(numbers, "Si", "No"),
              ifelse(symbols, "Si", "No")))
}


# Programa principal
main <- function() {
  repeat {
    interactive_generator()
    again <- tolower(readline(prompt = "\nGenerar mas contrase√±as? (s/n): "))
    if (!again %in% c("s", "si", "s√≠", "y", "yes")) {
      cat("\nHasta pronto!\n")
      break
    }
  }
}

# Ejecutar
main()
```

## 3. Adivina la palabra

Crea un peque√±o juego que consista en adivinar palabras en un n√∫mero m√°ximo de intentos:
- El juego comienza proponiendo una palabra aleatoria incompleta
  - Por ejemplo "m_ur_d_v", y el n√∫mero de intentos que le quedan
- El usuario puede introducir √∫nicamente una letra o una palabra (de la misma longitud que
  la palabra a adivinar)
  - Si escribe una letra y acierta, se muestra esa letra en la palabra. Si falla, se resta
    uno al n√∫mero de intentos
  - Si escribe una resoluci√≥n y acierta, finaliza el juego, en caso contrario, se resta uno
    al n√∫mero de intentos
  - Si el contador de intentos llega a 0, el jugador pierde
- La palabra debe ocultar de forma aleatoria letras, y nunca puede comenzar ocultando m√°s del 60%
- Puedes utilizar las palabras que quieras y el n√∫mero de intentos que consideres

```r
# Lista de palabras para el juego
words <- c(
  "python", "programacion", "algoritmo", "variable", "funcion",
  "biblioteca", "desarrollo", "computadora", "teclado", "pantalla",
  "internet", "servidor", "cliente", "archivo", "carpeta",
  "sistema", "memoria", "procesador", "software", "hardware",
  "elefante", "mariposa", "dinosaurio", "cocodrilo", "murcielago",
  "estrella", "universo", "galaxia", "planeta", "aventura"
)


hide_letters <- function(word, max_hidden_percent = 0.6) {
  #' Oculta letras aleatorias de una palabra.
  #' @param word Palabra a ocultar

  #' @param max_hidden_percent Porcentaje m√°ximo de letras a ocultar
  #' @return Lista con palabra oculta e √≠ndices ocultos

  chars <- unlist(strsplit(word, ""))
  length_word <- length(chars)
  max_hidden <- floor(length_word * max_hidden_percent)
  num_hidden <- sample(1:max_hidden, 1)

  # Seleccionar √≠ndices aleatorios para ocultar
  hidden_indices <- sample(1:length_word, num_hidden)

  # Crear palabra oculta
  hidden_chars <- chars
  hidden_chars[hidden_indices] <- "_"

  return(list(
    hidden_word = paste(hidden_chars, collapse = ""),
    hidden_indices = hidden_indices
  ))
}


display_status <- function(hidden_word, attempts, guessed_letters) {
  #' Muestra el estado actual del juego.

  chars <- unlist(strsplit(hidden_word, ""))

  cat("\n")
  cat(strrep("=", 50), "\n")
  cat("  Palabra:", paste(chars, collapse = " "), "\n")
  cat("  Intentos restantes:", strrep("* ", attempts), "(", attempts, ")\n")
  if (length(guessed_letters) > 0) {
    cat("  Letras usadas:", paste(sort(guessed_letters), collapse = ", "), "\n")
  }
  cat(strrep("=", 50), "\n")
}


play_game <- function() {
  #' Ejecuta una partida del juego.

  # Seleccionar palabra aleatoria
  word <- tolower(sample(words, 1))
  result <- hide_letters(word)
  hidden_word <- result$hidden_word

  # Configuraci√≥n
  attempts <- 6
  guessed_letters <- character(0)
  current_display <- unlist(strsplit(hidden_word, ""))
  word_chars <- unlist(strsplit(word, ""))
  
  cat("\n")
  cat(strrep("*", 50), "\n")
  cat("    JUEGO DE ADIVINAR PALABRAS\n")
  cat(strrep("*", 50), "\n")
  cat("\nREGLAS:\n")
  cat("  - Escribe una LETRA para revelar posiciones\n")
  cat("  - Escribe la PALABRA COMPLETA para resolver\n")
  cat("  - Si fallas, pierdes un intento\n")
  cat("\nLa palabra tiene", length(word_chars), "letras\n")

  while (attempts > 0) {
    display_status(
      paste(current_display, collapse = ""), attempts, guessed_letters
    )

    # Verificar si ya se adivin√≥
    if (!"_" %in% current_display) {
      cat("\n*** FELICIDADES! Has adivinado la palabra! ***\n")
      cat("    La palabra era:", toupper(word), "\n")
      return(TRUE)
    }

    # Obtener entrada del usuario
    guess <- tolower(trimws(readline(prompt = "\nTu letra o palabra: ")))

    if (guess == "") {
      cat("Error: Debes escribir algo.\n")
      next
    }

    guess_chars <- unlist(strsplit(guess, ""))

    # Caso: Una letra
    if (length(guess_chars) == 1) {
      if (!grepl("^[a-z]$", guess)) {
        cat("Error: Solo se permiten letras.\n")
        next
      }

      if (guess %in% guessed_letters) {
        cat("Ya usaste la letra '", guess, "'. Prueba otra.\n", sep = "")
        next
      }

      guessed_letters <- c(guessed_letters, guess)

      # Verificar si la letra est√° en la palabra
      if (guess %in% word_chars) {
        # Revelar todas las ocurrencias de la letra
        found <- FALSE
        for (i in seq_along(word_chars)) {
          if (word_chars[i] == guess && current_display[i] == "_") {
            current_display[i] <- guess
            found <- TRUE
          }
        }

        if (found) {
          cat("Bien! La letra '", guess, "' esta en la palabra.\n", sep = "")
        } else {
          cat("La letra '", guess, "' ya estaba visible.\n", sep = "")
        }
      } else {
        attempts <- attempts - 1
        cat("La letra '", guess, "' no esta en la palabra.\n", sep = "")
      }

      # Caso: Palabra completa
    } else if (length(guess_chars) == length(word_chars)) {
      if (guess == word) {
        cat("\n*** EXCELENTE! Adivinaste la palabra! ***\n")
        cat("    La palabra era:", toupper(word), "\n")
        return(TRUE)
      } else {
        attempts <- attempts - 1
        cat("'", guess, "' no es la palabra correcta.\n", sep = "")
      }

    } else {
      cat("La palabra debe tener ", length(word_chars),
          " letras, no ", length(guess_chars), ".\n", sep = "")
    }
  }

  # Sin intentos
  cat("\n")
  cat(strrep("=", 50), "\n")
  cat("GAME OVER! Te quedaste sin intentos.\n")
  cat("La palabra era:", toupper(word), "\n")
  cat(strrep("=", 50), "\n")
  FALSE
}


main <- function() {
  #' Funci√≥n principal del juego.

  wins <- 0
  losses <- 0

  repeat {
    result <- play_game()
    if (result) {
      wins <- wins + 1
    } else {
      losses <- losses + 1
    }

    cat("\nEstad√≠sticas:", wins, "victorias |", losses, "derrotas\n")

    again <- tolower(trimws(readline(prompt = "\nJugar otra vez? (s/n): ")))
    if (!again %in% c("s", "si", "s√≠", "y", "yes")) {
      cat("\nGracias por jugar! Hasta pronto!\n")
      break
    }
  }
}


# Ejecutar el juego
main()
```

```r
#' Detector del C√≥digo Konami
#' Secuencia: ‚Üë ‚Üë ‚Üì ‚Üì ‚Üê ‚Üí ‚Üê ‚Üí B A
#'
#' En R usamos teclas alternativas ya que no hay soporte nativo para flechas:
#' W=Arriba, S=Abajo, A=Izquierda, D=Derecha, B, A
#' Secuencia: W W S S A D A D B A

# Secuencia del C√≥digo Konami (adaptada para teclado)
KONAMI_CODE <- c("w", "w", "s", "s", "a", "d", "a", "d", "b", "a")

# S√≠mbolos para mostrar
SYMBOLS <- c(
  "w" = "‚Üë", "s" = "‚Üì", "a" = "‚Üê", "d" = "‚Üí",
  "b" = "B", "a" = "A"
)


show_welcome <- function() {
  cat("\n")
  cat(strrep("=", 60), "\n")
  cat("        DETECTOR DEL C√ìDIGO KONAMI\n")
  cat(strrep("=", 60), "\n")
  cat("\nSecuencia original: UP UP DOWN DOWN LEFT RIGHT LEFT RIGHT B A\n")
  cat("\nTeclas a usar (ya que R no detecta flechas):\n")
  cat("   W=Arriba, S=Abajo, A=Izquierda, D=Derecha\n")
  cat("\nSecuencia a introducir:\n")
  cat("   W W S S A D A D B A\n")
  cat("\nEscribe 'salir' para terminar\n")
  cat(strrep("-", 60), "\n\n")
}


show_easter_egg <- function() {
  cat("\n")
  cat(strrep("*", 60), "\n")
  cat("
  ‚ñà‚ñà‚ïó  ‚ñà‚ñà‚ïó ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó ‚ñà‚ñà‚ñà‚ïó   ‚ñà‚ñà‚ïó ‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó ‚ñà‚ñà‚ñà‚ïó   ‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïó
  ‚ñà‚ñà‚ïë ‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ïó  ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ïó ‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë
  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù ‚ñà‚ñà‚ïë   ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ñà‚ñà‚ïó ‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ñà‚ñà‚ñà‚ñà‚ïî‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë
  ‚ñà‚ñà‚ïî‚ïê‚ñà‚ñà‚ïó ‚ñà‚ñà‚ïë   ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë
  ‚ñà‚ñà‚ïë  ‚ñà‚ñà‚ïó‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë ‚ïö‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë  ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë ‚ïö‚ïê‚ïù ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë
  ‚ïö‚ïê‚ïù  ‚ïö‚ïê‚ïù ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù ‚ïö‚ïê‚ïù  ‚ïö‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïù  ‚ïö‚ïê‚ïù‚ïö‚ïê‚ïù     ‚ïö‚ïê‚ïù‚ïö‚ïê‚ïù
  \n")
  cat("        *** C√ìDIGO KONAMI ACTIVADO! ***\n")
  cat("        +30 VIDAS DESBLOQUEADAS\n")
  cat(strrep("*", 60), "\n\n")
}


display_progress <- function(sequence) {
  if (length(sequence) == 0) return()

  # Convertir a s√≠mbolos
  symbols <- sapply(sequence, function(k) {
    if (k %in% names(SYMBOLS)) SYMBOLS[k] else k
  })

  remaining <- length(KONAMI_CODE) - length(sequence)

  cat("   Secuencia: ", paste(symbols, collapse = " "))
  if (remaining > 0) {
    cat(" ... (", remaining, " restantes)\n", sep = "")
  } else {
    cat("\n")
  }
}


check_sequence <- function(input, current_sequence) {
  #' Verifica y actualiza la secuencia
  #' @return Lista con nueva secuencia y si se complet√≥

  chars <- tolower(unlist(strsplit(input, "")))
  completed <- FALSE

  for (char in chars) {
    # Ignorar caracteres no v√°lidos
    if (!char %in% c("w", "s", "a", "d", "b")) {
      next
    }

    expected_index <- length(current_sequence) + 1

    if (expected_index <= length(KONAMI_CODE) &&
          char == KONAMI_CODE[expected_index]) {
      current_sequence <- c(current_sequence, char)

      # Verificar si se complet√≥
      if (length(current_sequence) == length(KONAMI_CODE)) {
        completed <- TRUE
        break
      }
    } else {
      # Reiniciar secuencia
      if (length(current_sequence) > 0) {
        cat("   X Secuencia incorrecta. Reiniciando...\n")
      }
      current_sequence <- character(0)

      # Verificar si es inicio de nueva secuencia
      if (char == KONAMI_CODE[1]) {
        current_sequence <- c(char)
      }
    }
  }

  list(sequence = current_sequence, completed = completed)
}


main <- function() {
  show_welcome()

  current_sequence <- character(0)

  repeat {
    # Mostrar progreso actual
    remaining <- length(KONAMI_CODE) - length(current_sequence)
    prompt <- sprintf("Introduce tecla(s) [%d/%d]: ",
                      length(current_sequence), length(KONAMI_CODE))

    input <- readline(prompt = prompt)

    # Salir
    if (tolower(trimws(input)) == "salir") {
      cat("\nHasta pronto!\n")
      break
    }

    # Verificar secuencia
    result <- check_sequence(input, current_sequence)
    current_sequence <- result$sequence

    # Mostrar progreso
    display_progress(current_sequence)

    # Si se complet√≥ el c√≥digo
    if (result$completed) {
      show_easter_egg()
      current_sequence <- character(0)
      cat("Puedes intentarlo de nuevo o escribir 'salir'\n\n")
    }
  }
}


# Versi√≥n alternativa que lee tecla por tecla
main_single_key <- function() {
  show_welcome()
  cat("MODO: Una tecla a la vez\n\n")

  current_sequence <- character(0)

  repeat {
    # remaining <- length(KONAMI_CODE) - length(current_sequence)

    if (length(current_sequence) == 0) {
      cat("Esperando secuencia...\n")
    }

    input <- readline(prompt = "> ")

    # Salir
    if (tolower(trimws(input)) == "salir") {
      cat("\nHasta pronto!\n")
      break
    }

    # Tomar solo el primer car√°cter
    char <- tolower(substr(trimws(input), 1, 1))

    # Ignorar entrada vac√≠a o caracteres no v√°lidos
    if (char == "" || !char %in% c("w", "s", "a", "d", "b", "a")) {
      cat("Teclas validas: W, S, A, D, B\n")
      next
    }

    expected_index <- length(current_sequence) + 1

    if (expected_index <= length(KONAMI_CODE) && 
          char == KONAMI_CODE[expected_index]) {
      current_sequence <- c(current_sequence, char)
      display_progress(current_sequence)

      # Si se complet√≥
      if (length(current_sequence) == length(KONAMI_CODE)) {
        show_easter_egg()
        current_sequence <- character(0)
        cat("Puedes intentarlo de nuevo o escribir 'salir'\n\n")
      }
    } else {
      if (length(current_sequence) > 0) {
        cat("   X Secuencia incorrecta. Reiniciando...\n")
      }
      current_sequence <- character(0)

      if (char == KONAMI_CODE[1]) {
        current_sequence <- c(char)
        display_progress(current_sequence)
      }
    }
  }
}


# Ejecutar
cat("Selecciona modo:\n")
cat("1. Escribir varias teclas a la vez (ej: 'wwssadad')\n")
cat("2. Una tecla a la vez\n")
mode <- readline(prompt = "Opci√≥n (1/2): ")

if (mode == "2") {
  main_single_key()
} else {
  main()
}
```

## 5. Teclado T9

Los primeros dispositivos m√≥viles ten√≠an un teclado llamado T9
con el que se pod√≠a escribir texto utilizando √∫nicamente su
teclado num√©rico (del 0 al 9).
 Crea una funci√≥n que transforme las pulsaciones del T9 a su
representaci√≥n con letras.
- Debes buscar cu√°l era su correspondencia original.
- Cada bloque de pulsaciones va separado por un gui√≥n.
- Si un bloque tiene m√°s de un n√∫mero, debe ser siempre el mismo.
- Ejemplo:
    Entrada: 6-666-88-777-33-3-33-888
    Salida: MOUREDEV

```r
#' Decodificador T9 Multi-tap
#' Convierte pulsaciones de teclado num√©rico a letras.
#'
#' Correspondencia T9:
#'     1: .    2: ABC   3: DEF
#'     4: GHI  5: JKL   6: MNO
#'     7: PQRS 8: TUV   9: WXYZ
#'     0: (espacio)

# Mapeo T9: cada n√∫mero corresponde a un grupo de letras
T9_MAP <- list(
  "1" = ".",
  "2" = "ABC",
  "3" = "DEF",
  "4" = "GHI",
  "5" = "JKL",
  "6" = "MNO",
  "7" = "PQRS",
  "8" = "TUV",
  "9" = "WXYZ",
  "0" = " "
)


decode_t9 <- function(input_code) {
  #' Decodifica una secuencia T9 a texto.
  #' @param input_code Secuencia de pulsaciones separadas por guiones
  #' @return Texto decodificado en may√∫sculas

  if (is.null(input_code) || trimws(input_code) == "") {
    return("")
  }

  # Separar bloques por guiones
  blocks <- unlist(strsplit(trimws(input_code), "-"))
  result <- character(0)

  for (block in blocks) {
    block <- trimws(block)

    if (block == "") next

    # Validar que todos los caracteres sean n√∫meros
    if (!grepl("^[0-9]+$", block)) {
      stop(paste0("Bloque invalido: '", block, "'. Solo se permiten n√∫meros."))
    }

    # Validar que todos los d√≠gitos sean iguales
    chars <- unlist(strsplit(block, ""))
    if (length(unique(chars)) > 1) {
      stop(paste0(
        "Bloque invalido: '", block, "'. Todos los d√≠gitos deben ser iguales."
      ))
    }

    digit <- chars[1]
    count <- length(chars)

    if (!digit %in% names(T9_MAP)) {
      stop(paste0("D√≠gito no valido: '", digit, "'"))
    }

    letters <- T9_MAP[[digit]]
    letters_vec <- unlist(strsplit(letters, ""))

    # Calcular √≠ndice (con ciclo si se excede)
    index <- ((count - 1) %% length(letters_vec)) + 1
    result <- c(result, letters_vec[index])
  }

  paste(result, collapse = "")
}


encode_t9 <- function(text) {
  #' Codifica texto a secuencia T9.
  #' @param text Texto a codificar
  #' @return Secuencia T9 separada por guiones

  # Crear mapeo inverso
  reverse_map <- list()
  for (digit in names(T9_MAP)) {
    letters <- unlist(strsplit(T9_MAP[[digit]], ""))
    for (i in seq_along(letters)) {
      reverse_map[[letters[i]]] <- paste(rep(digit, i), collapse = "")
    }
  }

  result <- character(0)
  chars <- unlist(strsplit(toupper(text), ""))

  for (char in chars) {
    if (char %in% names(reverse_map)) {
      result <- c(result, reverse_map[[char]])
    } else if (char == " ") {
      result <- c(result, "0")
    }
    # Ignorar caracteres no mapeables
  }

  paste(result, collapse = "-")
}


show_keyboard <- function() {
  cat("
    +===================================+
    |        TECLADO T9 MULTI-TAP       |
    +===========+===========+===========+
    |  1  .     |  2  ABC   |  3  DEF   |
    +===========+===========+===========+
    |  4  GHI   |  5  JKL   |  6  MNO   |
    +===========+===========+===========+
    |  7  PQRS  |  8  TUV   |  9  WXYZ  |
    +===========+===========+===========+
    |     *     |  0  [ ]   |     #     |
    +===========+===========+===========+
  \n")
}


interactive_mode <- function() {
  cat("\n")
  cat(strrep("=", 50), "\n")
  cat("       DECODIFICADOR T9 MULTI-TAP\n")
  cat(strrep("=", 50), "\n")

  show_keyboard()

  cat("INSTRUCCIONES:\n")
  cat("  - Cada bloque de pulsaciones se separa con '-'\n")
  cat("  - Un bloque debe contener el mismo numero repetido\n")
  cat("  - Ejemplo: 6-666-88-777-33-3-33-888 = MOUREDEV\n")
  cat(strrep("-", 50), "\n")

  repeat {
    cat("\nQue deseas hacer?\n")
    cat("  1. Decodificar T9 -> Texto\n")
    cat("  2. Codificar Texto -> T9\n")
    cat("  3. Salir\n")

    option <- trimws(readline(prompt = "\nOpci√≥n: "))

    if (option == "1") {
      code <- trimws(readline(prompt = "\nIntroduce c√≥digo T9: "))
      tryCatch({
        result <- decode_t9(code)
        cat("\n*** C√≥digo T9:", code, "***\n")
        cat("\n*** Resultado:", result, "***\n")
      }, error = function(e) {
        cat("\nError:", e$message, "\n")
      })

    } else if (option == "2") {
      text <- trimws(readline(prompt = "\nIntroduce texto: "))
      result <- encode_t9(text)
      cat("\n*** Texto:", text, "***\n")
      cat("\n*** C√≥digo T9:", result, "***\n")

    } else if (option == "3") {
      cat("\nHasta pronto!\n")
      break

    } else {
      cat("Opci√≥n no valida\n")
    }
  }
}

interactive_mode()
```

# 6. La carrera de coches

Crea un programa que simule la competici√≥n de dos coches en una pista.
- Los dos coches estar√°n representados por üöô y üöó. Y la meta por üèÅ.
- Cada pista tendr√° entre 1 y 3 √°rboles üå≤ colocados de forma aleatoria.
- Las dos pistas tendr√°n una longitud configurable de guiones bajos "_".
- Los coches comenzar√°n en la parte derecha de las pistas. Ejemplo:
  üèÅ____üå≤_____üöô
  üèÅ_üå≤____üå≤___üöó

El juego se desarrolla por turnos de forma autom√°tica, y cada segundo
se realiza una acci√≥n sobre los coches (movi√©ndose a la vez), hasta que
uno de ellos (o los dos a la vez) llega a la meta.
- Acciones:
  - Avanzar entre 1 a 3 posiciones hacia la meta.
  - Si al avanzar, el coche finaliza en la posici√≥n de un √°rbol,
    se muestra üí• y no avanza durante un turno.
  - Cada turno se imprimen las pistas y sus elementos.
  - Cuando la carrera finalice, se muestra el coche ganador o el empate.

```r
#' Carrera de Coches
#' Simulaci√≥n de una carrera entre dos coches con obst√°culos

# Clase Car usando listas
create_car <- function(name, emoji, track_length) {
  list(
    name = name,
    emoji = emoji,
    position = track_length,  # Empieza al final
    track_length = track_length,
    stunned = FALSE,  # Si choc√≥ con un √°rbol
    finished = FALSE
  )
}


move_car <- function(car, trees) {
  #' Mueve el coche
  #' @return Lista con car actualizado, posiciones avanzadas y si choc√≥

  if (car$stunned) {
    car$stunned <- FALSE
    return(list(car = car, advance = 0, crashed = FALSE))
  }

  if (car$finished) {
    return(list(car = car, advance = 0, crashed = FALSE))
  }

  # Avanzar entre 1 y 3 posiciones
  advance <- sample(1:3, 1)
  new_position <- max(0, car$position - advance)

  # Verificar si lleg√≥ a la meta
  if (new_position == 0) {
    car$position <- 0
    car$finished <- TRUE
    return(list(car = car, advance = advance, crashed = FALSE))
  }

  # Verificar si choc√≥ con un √°rbol
  if (new_position %in% trees) {
    car$position <- new_position
    car$stunned <- TRUE
    return(list(car = car, advance = advance, crashed = TRUE))
  }

  car$position <- new_position
  list(car = car, advance = advance, crashed = FALSE)
}


# Clase RaceTrack usando listas
create_track <- function(length = 20) {
  list(
    length = length,
    trees = place_trees(length)
  )
}


place_trees <- function(length) {
  #' Coloca entre 1 y 3 √°rboles aleatoriamente
  num_trees <- sample(1:3, 1)
  available_positions <- 1:(length - 1)  # No en meta (0) ni inicio (length)
  sample(available_positions, min(num_trees, length(available_positions)))
}


display_track <- function(track, car) {
  #' Genera la visualizaci√≥n de la pista

  track_viz <- rep("_", track$length + 1)
  track_viz[1] <- "[M]"  # Meta (usando [M] en lugar de emoji)

  # Colocar √°rboles
  for (tree_pos in track$trees) {
    track_viz[tree_pos + 1] <- "[T]"  # +1 porque R usa √≠ndices 1-based
  }

  # Colocar coche
  car_idx <- car$position + 1
  if (car$stunned && car$position %in% track$trees) {
    track_viz[car_idx] <- "[X]"  # Choque
  } else if (car$position == 0) {
    track_viz[car_idx] <- paste0(car$emoji, "[M]")
  } else {
    track_viz[car_idx] <- car$emoji
  }

  paste(track_viz, collapse = "")
}


clear_screen <- function() {
  if (.Platform$OS.type == "windows") shell("cls") else system("clear")
}


display_race <- function(car1, track1, car2, track2, turn) {
  #' Muestra el estado actual de la carrera

  clear_screen()
  cat("\n")
  cat(strrep("=", 60), "\n")
  cat("                  CARRERA DE COCHES\n")
  cat(strrep("=", 60), "\n")
  cat("\nTurno:", turn, "\n\n")
  cat(car1$name, car1$emoji, "\n")
  cat(display_track(track1, car1), "\n\n")
  cat(car2$name, car2$emoji, "\n")
  cat(display_track(track2, car2), "\n")
  cat(strrep("-", 60), "\n")
}


show_action <- function(car, advance, crashed) {
  #' Muestra la acci√≥n realizada por un coche

  if (crashed) {
    cat("[X]", car$name, "choco con un √°rbol y queda aturdido!\n")
  } else if (advance > 0 && !car$finished) {
    cat("[>]", car$name, "avanza", advance, "posici√≥n(es)\n")
  } else if (car$stunned) {
    cat("[~]", car$name, "esta recuper√°ndose del choque\n")
  } else if (car$finished && advance > 0) {
    cat("[!]", car$name, "llego a la meta!\n")
  }
}


race_game <- function(track_length = 20) {
  #' Ejecuta la carrera de coches

  # Crear pistas
  track1 <- create_track(track_length)
  track2 <- create_track(track_length)

  # Crear coches
  car1 <- create_car("Coche 1", "[C1]", track_length)
  car2 <- create_car("Coche 2", "[C2]", track_length)

  turn <- 0

  # Mostrar estado inicial
  display_race(car1, track1, car2, track2, turn)
  cat("\nLa carrera esta por comenzar!\n")
  invisible(readline(prompt = "\nPresiona Enter para iniciar..."))

  # Loop principal
  while (!(car1$finished || car2$finished)) {
    turn <- turn + 1
    Sys.sleep(1)  # Esperar 1 segundo

    # Mover coches
    result1 <- move_car(car1, track1$trees)
    car1 <- result1$car
    advance1 <- result1$advance
    crashed1 <- result1$crashed

    result2 <- move_car(car2, track2$trees)
    car2 <- result2$car
    advance2 <- result2$advance
    crashed2 <- result2$crashed

    # Mostrar estado
    display_race(car1, track1, car2, track2, turn)

    # Mostrar acciones
    show_action(car1, advance1, crashed1)
    show_action(car2, advance2, crashed2)

    # Verificar fin de carrera
    if (car1$finished || car2$finished) {
      break
    }
  }

  # Mostrar resultado
  Sys.sleep(1)
  cat("\n")
  cat(strrep("=", 60), "\n")
  cat("                  RESULTADO FINAL\n")
  cat(strrep("=", 60), "\n")

  if (car1$finished && car2$finished) {
    cat("\n[=] EMPATE! Ambos coches llegaron a la meta\n")
  } else if (car1$finished) {
    cat("\n[*]", car1$name, car1$emoji, "GANA LA CARRERA!\n")
  } else {
    cat("\n[*]", car2$name, car2$emoji, "GANA LA CARRERA!\n")
  }

  cat("\nTurnos totales:", turn, "\n")
  cat(strrep("=", 60), "\n")
}


main <- function() {
  #' Funci√≥n principal

  cat("\n")
  cat(strrep("=", 60), "\n")
  cat("                  CARRERA DE COCHES\n")
  cat(strrep("=", 60), "\n")
  cat("\n[!] Dos coches compiten en pistas con obst√°culos\n")
  cat("[T] Los arboles pueden hacer que los coches choquen\n")
  cat("[M] El primero en llegar a la meta gana\n\n")
  cat("Leyenda: [M]=Meta, [T]=√Årbol, [X]=Choque, [C1/C2]=Coches\n\n")

  repeat {
    length_input <- readline(
      prompt = "Longitud de la pista (10-50, recomendado 20): "
    )
    length <- suppressWarnings(as.integer(length_input))

    if (!is.na(length) && length >= 10 && length <= 50) {
      break
    }
    cat("Error: La longitud debe estar entre 10 y 50\n")
  }

  repeat {
    race_game(length)

    again <- tolower(trimws(readline(prompt = "\nOtra carrera? (s/n): ")))
    if (!again %in% c("s", "si", "s√≠", "y", "yes")) {
      cat("\nGracias por jugar!\n")
      break
    }
  }
}

# Ejecutar
main()
```
